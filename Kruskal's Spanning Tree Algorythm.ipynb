```python
class Node(object):
    def __init__(self, name):
        self.name = name;
        self.parent = self;
        self.rank = 0;
    
    def find(self):
        if self.parent != self:
            self.parent = self.parent.find();
        return self.parent;
    
    def merge(self, other):
        root1 = self.find();
        root2 = other.find();
        if root1 == root2:
            return False;
        if root1.rank > root2.rank:
            root2.parent = root1;
            
        elif root1.rank < root2.rank:
            root1.parent = root2;
        else:
            root1.parent = root2;
            root2.rank = root2.rank + 1;
        return True;
    
    
class Vertex(object):
    def __init__(self, name):
        self.name = name;
        self.node = Node(name);

class Edge(object):
    
    def __init__(self, weight, startVertex, endVertex):
        self.startVertex = startVertex;
        self.endVertex = endVertex;
        self.weight = weight;
        
    def __cmp__(self, other):
        return self.smp(self.weight, other.weigh);
    
    def __lt__(self, other):
        return self.weight < other.weight;

    
    
    
    
    
    
```
